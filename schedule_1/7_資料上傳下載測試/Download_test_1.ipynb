```python
#載入模組
import cv2
import numpy as np
import struct
import MySQLdb
```

```python
#連接資料庫
db = MySQLdb.connect(
    host="localhost", #伺服器位址
    user="root", #使用者名稱
    passwd="", #使用者密碼
    db="test", #資料庫名稱
    charset="utf8" #設定編碼
)
cursor = db.cursor() #資料整合為指標
```

```python
#轉換回圖片
def Change(_results,_i) :
   nparr = np.frombuffer(_results[_i][0], np.uint8)  #轉換回陣列
   img_decode = cv2.imdecode(nparr, cv2.IMREAD_COLOR)  #將數據編碼回img
   return img_decode
```

```python
cursor.execute("SELECT img FROM `img`") #SQL語法 //搜尋 date資料表中的資料
```

```python
results = cursor.fetchall() #接收回傳值
```

```python
def printimg (_imagePath):
   lena = cv2.cvtColor(_imagePath, cv2.COLOR_BGR2RGB) #色彩校正
   plt.imshow(lena)
   plt.axis('off')
   plt.show()
```

```python
print("\n輸出結果\n\n")
for i in range(0,cursor.execute("SELECT img FROM `img`")):
   print("第",i+1,"張圖片")
   printimg(Change(results,i))
```
